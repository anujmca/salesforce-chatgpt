## Iteration Folds Results
The training script compared LightGBM, XGBoost, CatBoost, Random Forest, and Logistic Regression on each fold.
The results were saved to `processed_data/training_results.pkl`.

```python
import pickle
with open('processed_data/training_results.pkl', 'rb') as f:
    results = pickle.load(f)

for fold, algos in results.items():
    print(f'\n--- {fold} ---')
    for name, metrics in algos.items():
        print(f'{name}: ROC AUC = {metrics["ROC AUC"]:.4f}')
```
